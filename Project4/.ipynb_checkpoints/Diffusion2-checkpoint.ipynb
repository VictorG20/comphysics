# 2. Diffusion

We attempt to integrate numerically the equation
\begin{equation}
    \frac{\partial T(x, t)}{\partial t} = \frac{K}{C \rho} \frac{\partial^{2}T}{\partial x^{2}},
\end{equation}


# a) FTCS


The Forward Time Centered Space scheme (FTCS), as indicated by the name, uses a Forward Difference for time derivatives together with an Euler method (setting the function at time $t_{n}$) and a Central Finite Difference for spatial derivatives. Thus, for the diffusion equation we have

\begin{equation}
    \frac{\partial T}{\partial t} = F(t, x, \partial_{x}^{2}T) \implies \frac{\partial T}{\partial t} \overset{\text{F.D.}}{\approx} \frac{T(t_{n + 1}, x_{j}) - T(t_{n}, x_{j})}{\Delta t} \qquad \frac{T(t_{n + 1}, x_{j}) - T(t_{n}, x_{j})}{\Delta t} = F\left(t_{n}, x_{i}, \partial_{x}^{2}T(t_{n}, x_{i})\right) \quad \text{Euler method for time.}
\end{equation}

Then, for space, we use the given form of $F(t, x, \partial_{x}^{2}T)$ to obtain

\begin{equation}
    \frac{\partial^{2}T}{\partial x^{2}}(t_{n}, x_{j}) \overset{\text{C.F.D.}}{\approx} \frac{T(t_{n}, x_{j + 1}) - 2 T(t_{n}, x_{j}) + T(t_{n}, x_{j - 1})}{(\Delta x)^{2}}
\end{equation}

Hence, the equation discretized with a FTCS scheme reads

\begin{equation}
    \frac{T(t_{n + 1}, x_{j}) - T(t_{n}, x_{j})}{\Delta t} = \frac{K}{C \rho} \frac{T(t_{n}, x_{j + 1}) - 2 T(t_{n}, x_{j}) + T(t_{n}, x_{j - 1})}{(\Delta x)^{2}}.
\end{equation}

Thus, we can now obtain the value of the temperature at each point for every future time

\begin{equation}
    \boxed{T(t_{n + 1}, x_{j}) = T(t_{n}, x_{j}) + \frac{K}{C\rho}\frac{\Delta t}{(\Delta x)^{2}}\left(T(t_{n}, x_{j + 1}) - 2 T(t_{n}, x_{j}) + T(t_{n}, x_{j - 1})\right)}
\end{equation}

```julia
using Plots
using LinearAlgebra
using BenchmarkTools

function ftcs(T, a, u0 = 0., u_end=0.)
    in_points = size(T)[1]
    T2 = copy(T)
    
    T[1] = (1-2*a)*T2[1] + a*(T2[2] + u0)
    T[end] = (1-2*a)*T2[end] + a*(u_end + T2[end-1])
    for j ∈ 2:in_points-1
        T[j] = (1-2*a)*T2[j] + a*(T2[j+1] + T2[j-1])
    end
end



function diffusion(T_initial::Function, method::Function, Δx::Float64, Δt::Float64, nt::Int, sample,L, K, C, ρ)
    in_points = trunc(Int, L/Δx) - 1
    T = Float64[T_initial(x) for x in range(Δx, step=Δx, length=in_points)]
    
    sample_size = trunc(Int, nt/sample)
    if mod(nt, sample) == 0
        sample_size += 1 
        T_samples = zeros(in_points, sample_size)
        T_samples[:, 1] = copy(T)
    else
        T_samples = zeros(in_points, sample_size)
    end
    
    if method == ftcs
        a = K*Δt/(C*ρ*(Δx)^2)
    end
    
    for step ∈ 1:nt
        method(T, a)
        
        if mod(nt-step,sample) != 0
            continue
        end
        
        j = sample_size - ((nt-step)÷sample)
        T_samples[:,j] = copy(T)
    end
    return T_samples
end
```

```julia
Δx = 0.01
Δt = 0.1
nt = 10000

sample = 1

L = 1.
K = 210.
C = 900.
ρ = 2700.


T_initial(x) = sin(π*x/L)

temperatures = diffusion(T_initial, ftcs, Δx, Δt, nt, sample, L, K, C, ρ)
```

```julia
temperatures = transpose(temperatures)  # Needed to plot the time in the vertical axis
boundary_points = [0. for _ in range(1, size(temperatures)[1])]
temperatures = [boundary_points temperatures boundary_points]  # Add the boundary x=0 for aesthetics.

sample_size = trunc(Int, nt/sample)

x = range(0, 1, step=Δx)
t = range(0, step=0.1, length=10001)
temperature_heatmap = heatmap(x, t, temperatures, xlabel="\$ x \$", ylabel="\$ t\$", 
                              colorbar_title="\$T \\;(^{\\circ}{C})\$", dpi=600, labelfontsize=13)
```

```julia
@benchmark diffusion(T_initial, ftcs, Δx, Δt, nt, sample, L, K, C, ρ)
```

```julia
function heat_ftcs(T_initial::Function, Δx, Δt, nt, L, K, C, ρ)
    nx = trunc(Int, L/Δx) - 1  # Inner points of the grid
    T = Float64[T_initial(x) for x in range(Δx, step=Δx, length=nx)]
    
    
    a = K*Δt/(C*ρ*(Δx)^2)
    d = [1-2*a for _ ∈ 1:nx]
    sd = [a for _ ∈ 1:nx-1]
    A = Tridiagonal(sd, d, sd)
        
    for step ∈ 1:nt
       T[:] .= A*T[:]
    end

    return T
end
```

```julia
Δx = 0.01
Δt = 0.1
nt = 10000

L = 1.
K = 210.
C = 900.
ρ = 2700.


T_initial(x) = sin(π*x/L)

@benchmark heat_ftcs(T_initial::Function, Δx, Δt, nt, L, K, C, ρ)
```
